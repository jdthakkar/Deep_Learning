```python
img_list=tf.data.Dataset.list_files("/content/flower_photos/*/*/*")
```

```python
for img in img_list:
  print(img.numpy())
```

```python
img_count=len(img_list)
```

```python
train_size=int(img_count*0.8)
```

```python
train_ds=img_list.take(train_size)
test_ds=img_list.skip(train_size)
```

```python
def get_y(img_path):
  import os
  y=tf.strings.split(img_path,os.path.sep)[-2]
  return y
```

```python
def get_x(img_path):
  y=get_y(img_path)
  x=tf.io.read_file(img_path)
  x=tf.image.decode_jpeg(x)
  x=tf.image.resize(x,(224,224))
  return x,y
```

```python
train_ds=train_ds.map(get_x)
```

```python
for i,j in train_ds.take(3):
  print(i)
  print(j)
```

```python
def scale(img,label):
  return img/255,label
```

```python
train_ds=train_ds.map(scale)
```

```python
for i,j in train_ds:
  print(i)
  print(j)
```
