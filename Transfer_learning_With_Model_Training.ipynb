```python
import tensorflow as tf
from tensorflow import keras
import cv2
import os
import PIL
import numpy as np
import matplotlib.pyplot as plt
```

```python
import zipfile
zip=zipfile.ZipFile("/content/F.zip")
zip.extractall()
```

```python
daisy=[]
file_lst=os.listdir('/content/flower_photos/flower_photos/daisy')
for i in file_lst:
  daisy.append('/content/flower_photos/flower_photos/daisy/'+i)

dandelion=[]
file_lst=os.listdir('/content/flower_photos/flower_photos/dandelion')
for i in file_lst:
  dandelion.append('/content/flower_photos/flower_photos/dandelion/'+i)

roses=[]
file_lst=os.listdir('/content/flower_photos/flower_photos/roses')
for i in file_lst:
  roses.append('/content/flower_photos/flower_photos/roses/'+i)

sunflower=[]
file_lst=os.listdir('/content/flower_photos/flower_photos/sunflowers')
for i in file_lst:
  sunflower.append('/content/flower_photos/flower_photos/sunflowers/'+i)

tulips=[]
file_lst=os.listdir('/content/flower_photos/flower_photos/tulips')
for i in file_lst:
  tulips.append('/content/flower_photos/flower_photos/tulips/'+i)
```

```python
flower_labels={
    'roses' :0,
    'daisy' :1,
    'dandelion' :2,
    'sunflowers' :3,
    'tulip' : 4
}
```

```python
import tensorflow_hub as hub
import tf_keras as keras
```

```python
x,y=[],[]
for images in daisy:
  img=cv2.imread(images)
  resized_img=cv2.resize(img,(224,224))
  x.append(resized_img)
  y.append(1)

for images in dandelion:
  img=cv2.imread(images)
  resized_img=cv2.resize(img,(224,224))
  x.append(resized_img)
  y.append(2)

for images in roses:
  img=cv2.imread(images)
  resized_img=cv2.resize(img,(224,224))
  x.append(resized_img)
  y.append(0)

for images in sunflower:
  img=cv2.imread(images)
  resized_img=cv2.resize(img,(224,224))
  x.append(resized_img)
  y.append(3)

for images in tulips:
  img=cv2.imread(images)
  resized_img=cv2.resize(img,(224,224))
  x.append(resized_img)
  y.append(4)

x=np.array(x)
y=np.array(y)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)

x_train_scaled=x_train/255
x_test_scaled=x_test/255
```

```python
classification_model_without_last_layer=hub.KerasLayer("https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4",input_shape=(224,224,3),trainable=False)
```

```python
transfer_model=keras.Sequential([
    classification_model_without_last_layer,
    keras.layers.Dense(5)
])
```

```python
transfer_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

transfer_model.fit(x_train_scaled,y_train,epochs=5)
```

```python
transfer_model.evaluate(x_test_scaled,y_test)
```
