```python
import tensorflow as tf
from tensorflow import keras
```

```python
words_len=100
input_category=10
output_category=3

text_input_1=keras.Input(shape=(None,),name='Text_1')
text_input_2=keras.Input(shape=(None,),name='Text_2')
category_input=keras.Input(shape=(input_category,),name='Category')
L1=keras.layers.Embedding(words_len,64)(text_input_1)
L2=keras.layers.Embedding(words_len,64)(text_input_2)
L3=keras.layers.LSTM(128)(L1)
L4=keras.layers.LSTM(32)(L2)
L5=keras.layers.Concatenate()([L3,L4,category_input])
L6=keras.layers.Dense(1,name='Binary_output')(L5)
L7=keras.layers.Dense(output_category,name='Category_output')(L5)
```

```python
model=keras.Model([text_input_1,text_input_2,category_input],[L6,L7])
```

```python
model.compile(
    optimizer='adam',
    loss={
        'Binary_output':keras.losses.BinaryCrossentropy(from_logits=True),
        'Category_output':keras.losses.CategoricalCrossentropy(from_logits=True)
    },
    metrics=['accuracy']
)
```
