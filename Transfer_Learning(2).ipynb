```python
import tensorflow as tf
from tensorflow import keras
import PIL.Image as image
import cv2
import numpy as np
```

```python
model=tf.keras.applications.mobilenet_v2.MobileNetV2(input_shape=(224,224,3),include_top=False,weights='imagenet')
```

```python
model.trainable=False
```

```python
model2=keras.Sequential([
    model,
    keras.layers.Flatten(),
    keras.layers.Dense(5)
])
```

```python
import os
daisy=[]
file_lst=os.listdir('/content/flower_photos/flower_photos/daisy')
for i in file_lst:
  daisy.append('/content/flower_photos/flower_photos/daisy/'+i)

dandelion=[]
file_lst=os.listdir('/content/flower_photos/flower_photos/dandelion')
for i in file_lst:
  dandelion.append('/content/flower_photos/flower_photos/dandelion/'+i)

roses=[]
file_lst=os.listdir('/content/flower_photos/flower_photos/roses')
for i in file_lst:
  roses.append('/content/flower_photos/flower_photos/roses/'+i)

sunflower=[]
file_lst=os.listdir('/content/flower_photos/flower_photos/sunflowers')
for i in file_lst:
  sunflower.append('/content/flower_photos/flower_photos/sunflowers/'+i)

tulips=[]
file_lst=os.listdir('/content/flower_photos/flower_photos/tulips')
for i in file_lst:
  tulips.append('/content/flower_photos/flower_photos/tulips/'+i)
```

```python
flower_labels={
    'roses' :0,
    'daisy' :1,
    'dandelion' :2,
    'sunflowers' :3,
    'tulip' : 4
}
```

```python
x,y=[],[]
for images in daisy:
  img=cv2.imread(images)
  resized_img=cv2.resize(img,(224,224))
  x.append(resized_img)
  y.append(1)

for images in dandelion:
  img=cv2.imread(images)
  resized_img=cv2.resize(img,(224,224))
  x.append(resized_img)
  y.append(2)

for images in roses:
  img=cv2.imread(images)
  resized_img=cv2.resize(img,(224,224))
  x.append(resized_img)
  y.append(0)

for images in sunflower:
  img=cv2.imread(images)
  resized_img=cv2.resize(img,(224,224))
  x.append(resized_img)
  y.append(3)

for images in tulips:
  img=cv2.imread(images)
  resized_img=cv2.resize(img,(224,224))
  x.append(resized_img)
  y.append(4)

x=np.array(x)
y=np.array(y)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)

x_train_scaled=x_train/255
x_test_scaled=x_test/255
```

```python
model2.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model2.fit(x_train_scaled,y_train,epochs=25)
```
