```python
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
from matplotlib import pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.fashion_mnist.load_data()
```

```python
classes=[
    'T-shirt/top',
    'Trouser',
    'Pullover',
    'Dress',
    'Coat',
    'Sandal',
    'Shirt',
    'Sneaker',
    'Bag',
    'Ankle boot'
]
```

```python
x_train_normalized=x_train/255
x_test_normalized=x_test/255
```

```python
input_layer=tf.keras.Input(shape=(28,28))
```

```python
L1=layers.Flatten()(input_layer)
```

```python
L2=layers.Dense(128,activation='relu')(L1)
```

```python
output_layer=layers.Dense(10,activation='softmax')(L2)
```

```python
model=tf.keras.Model(inputs=input_layer,outputs=output_layer,name='Function_Model_1')
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x_train_normalized,y_train,epochs=10)
```

```python
model.evaluate(x_test_normalized,y_test)
```

```python
img=x_test[50]
pred=model.predict(img[np.newaxis,...])
```
