```python
from flask import Flask

app=Flask(__name__)


@app.route('/')

def root():
    return "This is not main directory"


@app.route('/potato')

def welcome():
    return "I like to eat potato"


@app.route('/yasso')

def yassin():
    return "Hello Yassin, sup man!!"


@app.route('/english')

def english():
    return "This is an english course"

if __name__=='__main__':
    app.run(debug=True)
```
